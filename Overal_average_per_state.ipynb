```python
import matplotlib.pyplot as plt
import pandas as pd
import pathlib as path
import numpy as np
```

```python
#Read path to DataFrame
path = pd.ExcelFile("Resources/total_sheltered.xlsx")
total_2013 = pd.read_excel(path, '2013')
total_2014 = pd.read_excel(path, '2014')
total_2015 = pd.read_excel(path, '2015')
total_2016 = pd.read_excel(path, '2016')
total_2017 = pd.read_excel(path, '2017')
total_2018 = pd.read_excel(path, '2018')
total_2019 = pd.read_excel(path, '2019')
total_2020 = pd.read_excel(path, '2020')
```

```python
#clean to Texas2013
total_2013.fillna('', inplace=True)
total13 = total_2013[total_2013['CoC Number'].str.startswith('TX')].reset_index(drop=True)

#clean to Texas2014
total_2014.fillna('', inplace=True)
total14 = total_2014[total_2014['CoC Number'].str.startswith('TX')].reset_index(drop=True)

#clean to Texas2015
total_2015.fillna('', inplace=True)
total15 = total_2015[total_2015['CoC Number'].str.startswith('TX')].reset_index(drop=True)

#clean to Texas2016
total_2016.fillna('', inplace=True)
total16 = total_2016[total_2016['CoC Number'].str.startswith('TX')].reset_index(drop=True)

#clean to Texas2017
total_2017.fillna('', inplace=True)
total17 = total_2017[total_2017['CoC Number'].str.startswith('TX')].reset_index(drop=True)

#clean to Texas2018
total_2018.fillna('', inplace=True)
total18 = total_2018[total_2018['CoC Number'].str.startswith('TX')].reset_index(drop=True)

#clean to Texas2019
total_2019.fillna('', inplace=True)
total19 = total_2019[total_2019['CoC Number'].str.startswith('TX')].reset_index(drop=True)

#clean to Texas2020
total_2020.fillna('', inplace=True)
total20 = total_2020[total_2020['CoC Number'].str.startswith('TX')].reset_index(drop=True)
total20
```

```python
# merge all clean tx DF together
t19 = pd.merge(total20, total19, on=['CoC Number', 'CoC Name'])
t18 = pd.merge(t19, total18, on=['CoC Number', 'CoC Name'])
t17 = pd.merge(t18, total17, on=['CoC Number', 'CoC Name'])
t16 = pd.merge(t17, total16, on=['CoC Number', 'CoC Name'])
t15 = pd.merge(t16, total15, on=['CoC Number', 'CoC Name'])
t14 = pd.merge(t15, total14, on=['CoC Number', 'CoC Name'])
t_years = pd.merge(t14, total13, on=['CoC Number', 'CoC Name'])
t_years.head()
```
