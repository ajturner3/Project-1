```python
# Select which columns to keep
columns_to_keep = ['State',
                   'Year',
                   'Overall Homeless - Non-Hispanic/Non-Latino',
                   'Overall Homeless - Hispanic/Latino',
                   'Overall Homeless - White',
                   'Overall Homeless - Black or African American',
                   'Overall Homeless - Asian',
                   'Overall Homeless - American Indian or Alaska Native',
                   'Overall Homeless - Native Hawaiian or Other Pacific Islander',
                   'Overall Homeless - Multiple Races',
                   'Sheltered Total Homeless - Non-Hispanic/Non-Latino',
                   'Sheltered Total Homeless - Hispanic/Latino',
                   'Sheltered Total Homeless - White',
                   'Sheltered Total Homeless - Black or African American',
                   'Sheltered Total Homeless - Asian',
                   'Sheltered Total Homeless - American Indian or Alaska Native',
                   'Sheltered Total Homeless - Native Hawaiian or Other Pacific Islander',
                   'Sheltered Total Homeless - Multiple Races']

```

```python
# Filter by rows (only CA and TX)
df_2020 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2020")
ca_tx_2020 = df_2020[(df_2020['State'] == 'CA') | (df_2020['State'] == 'TX')]
df_filtered_2020 = ca_tx_2020[columns_to_keep]
df_filtered_2020.head()
```

```python
# Filter by rows (only CA and TX)
df_2020 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2020")
ca_tx_2020 = df_2020[(df_2020['State'] == 'CA') | (df_2020['State'] == 'TX')]
df_filtered_2020 = ca_tx_2020[columns_to_keep]

df_2019 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2019")
ca_tx_2019 = df_2019[(df_2019['State'] == 'CA') | (df_2019['State'] == 'TX')]
df_filtered_2019 = ca_tx_2019[columns_to_keep]

df_2018 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2018")
ca_tx_2018 = df_2018[(df_2018['State'] == 'CA') | (df_2018['State'] == 'TX')]
df_filtered_2018 = ca_tx_2018[columns_to_keep]

df_2017 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2017")
ca_tx_2017 = df_2017[(df_2017['State'] == 'CA') | (df_2017['State'] == 'TX')]
df_filtered_2017 = ca_tx_2017[columns_to_keep]

df_2016 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2016")
ca_tx_2016 = df_2016[(df_2016['State'] == 'CA') | (df_2016['State'] == 'TX')]
df_filtered_2016 = ca_tx_2016[columns_to_keep]

df_2015 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2015")
ca_tx_2015 = df_2015[(df_2015['State'] == 'CA') | (df_2015['State'] == 'TX')]
df_filtered_2015 = ca_tx_2015[columns_to_keep]
```

```python
# Combine rows into a few datasheet
combined_data = pd.concat([df_filtered_2020, 
                           df_filtered_2019,
                           df_filtered_2018, 
                           df_filtered_2017, 
                           df_filtered_2016,
                           df_filtered_2015], ignore_index=True)

combined_data
```

```python
# Sort combined_data by like state
custom_order = ['CA', 'TX']
combined_data['State'] = pd.Categorical(combined_data['State'], categories=custom_order, ordered=True)
combined_data = combined_data.sort_values('State')
combined_data
```
